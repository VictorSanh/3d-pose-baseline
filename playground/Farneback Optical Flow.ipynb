```python
%matplotlib inline

import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
cap = cv2.VideoCapture("../data/playground/vtest.mp4")
```

```python
ret, frame1 = cap.read()
prvs = cv2.cvtColor(frame1, cv2.COLOR_BGR2GRAY)

hsv = np.zeros_like(frame1)
hsv[...,1] = 255
```

```python
f, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(30,30))

ax1.imshow(frame1)
ax1.set_title("BGR")

ax2.imshow(prvs)
ax2.set_title("BGR 2 GRAY")

ax3.imshow(hsv)
ax3.set_title("HSV")
```

```python
ret, frame1 = cap.read()
prvs = cv2.cvtColor(frame1, cv2.COLOR_BGR2GRAY)

ret, frame2 = cap.read()
next = cv2.cvtColor(frame2, cv2.COLOR_BGR2GRAY)

flow = cv2.calcOpticalFlowFarneback(prvs, next, None, 0.5, 3, 15, 3, 5, 1.2, 0)

mag, ang = cv2.cartToPolar(flow[...,0], flow[...,1])
hsv[...,0] = ang*180/np.pi/2
hsv[...,2] = cv2.normalize(mag,None,0,255,cv2.NORM_MINMAX)
bgr = cv2.cvtColor(hsv,cv2.COLOR_HSV2BGR)
#cv2.imshow('frame2',bgr)
```

```python
flow3Dim = np.zeros_like(frame1)
flow3Dim[:, :, 0] = flow[:, :, 0]
flow3Dim[:, :, 1] = flow[:, :, 1]

plt.imshow(flow3Dim)
plt.title("Optical Flow")
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(30,8))

im1 = ax1.imshow(mag)
ax1.set_title("mag")
fig.colorbar(im1, ax=ax1)

im2 = ax2.imshow(ang)
ax2.set_title("angle")
fig.colorbar(im2, ax=ax2)

fig.tight_layout()
```

```python
cap = cv2.VideoCapture("../data/playground/vtest2.mp4")

nbFrames = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))
print(nbFrames)

ret, frame1 = cap.read()
prvs = cv2.cvtColor(frame1, cv2.COLOR_BGR2GRAY)

i = 0

while(i<nbFrames):
    ret, frame2 = cap.read()
    next = cv2.cvtColor(frame2, cv2.COLOR_BGR2GRAY)

    flow = cv2.calcOpticalFlowFarneback(prvs, next, None, 0.5, 3, 15, 3, 5, 1.2, 0)
    print("Frame ", i, " and Frame ", i+1, " done")
    
    prvs = next
    
    i = i + 1
```
